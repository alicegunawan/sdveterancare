```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
```

```python
sail = pd.read_csv('sail_by_year.csv')
```

```python
# for easy viewing
sail_mental_health = sail.iloc[13:17]

benchmark_columns = [0] + list(range(3, 82, 5))
sandiego_columns = [0] + list(range(2, 82,5))
tenp_columns = [0] + list(range(4, 82, 5))
fiftyp_columns = [0] + list(range(5, 82, 5))
ninetyp_columns = [0] + list(range(6, 82, 5))

mental_health_benchmark = sail_mental_health.iloc[:, benchmark_columns]
mental_health_sandiego = sail_mental_health.iloc[:, sandiego_columns]
mental_health_tenp = sail_mental_health.iloc[:, tenp_columns]
mental_health_fiftyp = sail_mental_health.iloc[:, fiftyp_columns]
mental_health_ninetyp = sail_mental_health.iloc[:, ninetyp_columns]
sail_mental_health
```

```python
# san diego mental health
mental_health_sandiego_data = {"mental_health": list(mental_health_sandiego.iloc[0])[1:],\
                          "population_coverage": list(mental_health_sandiego.iloc[1])[1:],\
                          "continuity_of_care": list(mental_health_sandiego.iloc[2])[1:],\
                          "experience_of_care": list(mental_health_sandiego.iloc[3])[1:]}
mental_health_sandiego2 = pd.DataFrame(data = mental_health_sandiego_data)

# benchmark mental health
mental_health_benchmark_data = {"mental_health": list(mental_health_benchmark.iloc[0])[1:],\
                          "population_coverage": list(mental_health_benchmark.iloc[1])[1:],\
                          "continuity_of_care": list(mental_health_benchmark.iloc[2])[1:],\
                          "experience_of_care": list(mental_health_benchmark.iloc[3])[1:]}
mental_health_benchmark2 = pd.DataFrame(data = mental_health_benchmark_data)

# 10th percentile mental health
mental_health_tenp_data = {"mental_health": list(mental_health_tenp.iloc[0])[1:],\
                          "population_coverage": list(mental_health_tenp.iloc[1])[1:],\
                          "continuity_of_care": list(mental_health_tenp.iloc[2])[1:],\
                          "experience_of_care": list(mental_health_tenp.iloc[3])[1:]}
mental_health_tenp2 = pd.DataFrame(data = mental_health_tenp_data)

# 50th percentile mental health
mental_health_fiftyp_data = {"mental_health": list(mental_health_fiftyp.iloc[0])[1:],\
                          "population_coverage": list(mental_health_fiftyp.iloc[1])[1:],\
                          "continuity_of_care": list(mental_health_fiftyp.iloc[2])[1:],\
                          "experience_of_care": list(mental_health_fiftyp.iloc[3])[1:]}
mental_health_fiftyp2 = pd.DataFrame(data = mental_health_fiftyp_data)

# 90th percentile mental health
mental_health_ninetyp_data = {"mental_health": list(mental_health_ninetyp.iloc[0])[1:],\
                          "population_coverage": list(mental_health_ninetyp.iloc[1])[1:],\
                          "continuity_of_care": list(mental_health_ninetyp.iloc[2])[1:],\
                          "experience_of_care": list(mental_health_ninetyp.iloc[3])[1:]}
mental_health_ninetyp2 = pd.DataFrame(data = mental_health_ninetyp_data)
```

```python
pd.DataFrame({"mental_health_sandiego": list(mental_health_sandiego.iloc[0])[1:],\
                "mental_health_benchmark": list(mental_health_benchmark.iloc[0])[1:],\
                "mental_health_10thp": list(mental_health_tenp.iloc[0])[1:],\
                "mental_health_50thp": list(mental_health_fiftyp.iloc[0])[1:],\
                "mental_health_90thp": list(mental_health_ninetyp.iloc[0])[1:]}).plot()
```
